```python
def plot_reduced(X_reduced, y, title, sample_indices):
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='jet', alpha=0.5)
    plt.colorbar(scatter, ticks=[0, 1, 2], label='Species')
    plt.clim(-0.5, 2.5)
    plt.scatter(X_reduced[sample_indices, 0], X_reduced[sample_indices, 1], c='red', edgecolors='k', linewidths=2)
    plt.title(title)
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.show()
```
