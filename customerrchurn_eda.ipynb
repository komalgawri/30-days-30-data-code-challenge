```python
import pandas as pd
```

```python
df = pd.read_excel("dataset.xlsx")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
sns.set(style='whitegrid', palette='muted')
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(3, 2, 1)
sns.histplot(df['LoginFrequency'], kde=True, bins=30, color='blue')
plt.title('Login Frequency Distribution')

plt.subplot(3, 2, 2)
sns.histplot(df['UsageDuration'], kde=True, bins=30, color='green')
plt.title('Usage Duration Distribution')

plt.subplot(3, 2, 3)
sns.histplot(df['PurchaseHistory'], kde=True, bins=30, color='red')
plt.title('Purchase History Distribution')

plt.subplot(3, 2, 4)
sns.histplot(df['SubscriptionRenewals'], kde=True, bins=30, color='purple')
plt.title('Subscription Renewals Distribution')

plt.subplot(3, 2, 5)
sns.histplot(df['Age'], kde=True, bins=30, color='orange')
plt.title('Age Distribution')

plt.subplot(3, 2, 6)
sns.histplot(df['SatisfactionScores'], kde=True, bins=30, color='cyan')
plt.title('Satisfaction Scores Distribution')

plt.tight_layout()
plt.show()
plt.savefig('Numerical_Features_Distribution.png')
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(x='HistoricalChurn', y='LoginFrequency', data=df)
plt.title('Login Frequency by Churn Status')
plt.savefig('LoginFrequency_Boxplot.png')

plt.subplot(2, 2, 2)
sns.boxplot(x='HistoricalChurn', y='UsageDuration', data=df)
plt.title('Usage Duration by Churn Status')
plt.savefig('UsageDuration_Boxplot.png')

plt.subplot(2, 2, 3)
sns.boxplot(x='HistoricalChurn', y='PurchaseHistory', data=df)
plt.title('Purchase History by Churn Status')
plt.savefig('PurchaseHistory_Boxplot.png')

plt.subplot(2, 2, 4)
sns.boxplot(x='HistoricalChurn', y='SatisfactionScores', data=df)
plt.title('Satisfaction Scores by Churn Status')
plt.savefig('SatisfactionScores_Boxplot.png')

plt.tight_layout()
plt.savefig('Boxplots_by_Churn_Status.png')

plt.close()
```

```python
sns.pairplot(df.dropna(), hue='HistoricalChurn', diag_kind='kde')
plt.suptitle('Pairwise Relationships by Churn Status', y=1.02)
plt.savefig('Pairwise_Relationships.png')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.savefig('Missing_Values_Heatmap.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='HistoricalChurn', y='SupportTickets', data=df)
plt.title('Support Tickets by Churn Status')
plt.savefig('SupportTickets_Boxplot.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='HistoricalChurn', y='SatisfactionScores', data=df)
plt.title('Satisfaction Scores by Churn Status')
plt.savefig('SatisfactionScores_Violinplot.png')
plt.show()
```
