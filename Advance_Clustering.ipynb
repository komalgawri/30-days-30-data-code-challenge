```python
data = pd.read_csv("Mall_Customers.csv")
```

```python
data.head()
```

```python
features = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
X = features.to_numpy()
```

```python
import matplotlib.pyplot as plt


plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Data')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='plasma', s=50)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
from sklearn.cluster import MeanShift
```

```python
mean_shift = MeanShift(bandwidth=10)
mean_shift_labels = mean_shift.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=mean_shift_labels, cmap='plasma', s=50)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Mean Shift Clustering')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))


axes[0].scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='plasma', s=50)
axes[0].set_xlabel('Annual Income (k$)')
axes[0].set_ylabel('Spending Score (1-100)')
axes[0].set_title('DBSCAN Clustering')


axes[1].scatter(X[:, 0], X[:, 1], c=mean_shift_labels, cmap='plasma', s=50)
axes[1].set_xlabel('Annual Income (k$)')
axes[1].set_ylabel('Spending Score (1-100)')
axes[1].set_title('Mean Shift Clustering')

plt.show()
```
